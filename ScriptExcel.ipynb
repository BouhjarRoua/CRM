```python
import pandas as pd
from google.colab import files
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter

# Charger le fichier Excel contenant les données extraites d'Outlook
uploaded = files.upload()

# Charger le DataFrame à partir du fichier Excel téléchargé
for fn in uploaded.keys():
    df_contacts = pd.read_excel(fn)

# Charger le fichier Excel contenant le modèle/template
wb_template = load_workbook('/BRS Contact Template Test.xlsx')
ws_template = wb_template.active

# Récupérer les noms de colonnes du DataFrame
df_columns = set(df_contacts.columns)
print(f"df_columns {df_columns}")
# Récupérer les noms de colonnes du modèle Excel
template_columns = [cell.value for cell in ws_template[1]]
print(f"template_columns {template_columns}")
# Mapper les colonnes du DataFrame des contacts avec les champs du modèle
mapping = {
    'First Name': 'First Name',
    'Middle Name': 'Middle Name',
    'Last Name': 'Name',
    'Company': 'Company Name',
    'Department': 'Department',
    'Job Title': 'Role',
    'E-mail Address' : 'Email Address 1',
    'E-mail 2 Address' : 'Email Address 2',
    'Business Phone': ' Direct phone office',
    'Mobile Phone': 'Mobile 1',
    'Other Phone': 'Other Phone',
    'User 1': 'Ice',
    'Business Country/Region': 'Address 1 : Country',
    'Business Postal Code': 'Address 1: ZIP/Postal Code',
    'Business Street': 'Address : Street 1',
    # Ajoutez d'autres champs si nécessaire
}

for column in df_columns.intersection(set(mapping.keys())):
    # Convertir l'index de colonne en lettre de colonne dynamiquement
    column_letter = get_column_letter(template_columns.index(mapping[column]) + 1)
    print(column_letter)
    # Insérer les données dans la lettre de colonne dynamiquement
    for i, value in enumerate(df_contacts[column], start=2):
        ws_template[column_letter + str(i)] = value
        print(value)

# Sauvegarder le fichier Excel avec les données transférées
wb_template.save('contacts_template.xlsx')

# Télécharger le fichier résultant
files.download('contacts_template.xlsx')
```
